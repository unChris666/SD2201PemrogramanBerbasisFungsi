## 4. Apa yang dimaksud dengan pure functional programming language?

Pure Functional Programming Language adalah bahasa pemrograman yang mendasarkan semua implementasi pada paradigma fungsional. Dalam bahasa ini, setiap operasi program harus memiliki input yang benar dan output yang deterministik, tanpa efek samping. Fungsi harus didefinisikan sebagai relasi antara input dan output, dan tidak memiliki state atau data yang disimpan secara persisten. Ini membuat program dalam bahasa fungsional lebih mudah diprediksi dan diterapkan, serta mempermudah debugging dan testing.

## 6. Program Simple Password encryption

```python
def encrypt_password(password):
    result = ''
    for char in password:
        ascii_val = ord(char)
        first_val = (ascii_val // 26) + 80
        second_val = (ascii_val % 26) + 80
        third_val = '+' if first_val > second_val else '-'
        result += chr(first_val) + chr(second_val) + third_val
    return result

password = input("Enter the password: ")
if len(password) > 100:
    print("Password can not exceed 100 characters")
else:
    encrypted_password = encrypt_password(password)
    print("Encrypted password:", encrypted_password)
```
